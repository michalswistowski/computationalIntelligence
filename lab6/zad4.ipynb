```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

importy

```python
df = pd.read_csv('diabetes.csv')

df.describe().transpose()

df[['class']] = df[['class']].replace(['tested_negative', 'tested_positive'], [0, 1])
```

załadowanie danych\
zamiana wyników na 0,1

```python
target_column = ['class'] 
predictors = list(set(list(df.columns))-set(target_column))
df[predictors] = df[predictors]/df[predictors].max()
df.describe().transpose()

X = df[predictors].values
Y = df[target_column].values
```

rozdzielenie na predyktory i wyniki

```python
data_train, data_test, labels_train, labels_test = train_test_split(X, Y, test_size=0.30, random_state=40)
```

rozdzielenie na dane treningowe i testowe(70%,30%)

```python
model = Sequential()
model.add(Dense(6, input_shape=(8,), activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

stworzenie modelu:\
neurony wejściowe: 8\
pierwsza warstwa ukryta - 6 neuronów\
druga warstwa ukryta - 3 neurony\
aktywacje - relu


```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

kompilacja modelu - optimizer adam

```python
history = model.fit(data_train, labels_train, epochs=500, batch_size=10, validation_data=(data_test, labels_test))
```

wytrenowanie modelu na zbiorze treningowym

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model = Sequential()
model.add(Dense(6, input_shape=(8,), activation='sigmoid'))
model.add(Dense(3, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(data_train, labels_train, epochs=500, batch_size=10, validation_data=(data_test, labels_test))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
